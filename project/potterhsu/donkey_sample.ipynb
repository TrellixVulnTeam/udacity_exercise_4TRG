```python
import tensorflow as tf
from donkey import Donkey
%pylab inline
```

```python
image_batch, length_batch, digits_batch = Donkey.build_batch('../dataset/svhn/train.tfrecords', num_examples=100, batch_size=36, shuffled=False)
```

```python
sess = tf.InteractiveSession()
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)
```

```python
image_batch_val, length_batch_val, digits_batch_val = sess.run([image_batch, length_batch, digits_batch])
image_batch_val = (image_batch_val / 2.0) + 0.5
```
